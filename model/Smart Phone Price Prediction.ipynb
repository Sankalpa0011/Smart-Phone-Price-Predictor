```python
data = data.drop(columns=["num_cores","num_rear_cameras", "screen_size", "primary_camera_rear", "fast_charging", "os"], axis=1)
```

```python
data = data.dropna(how="any")
data.shape
```

```python
le = LabelEncoder()
data_encoded = data.apply(lambda x: le.fit_transform(x) if x.dtype == 'object' else x)
```

```python
data_encoded.corr()
```

```python
data_encoded.head()
```

```python
x = data_encoded.drop("price", axis=1)
y = data_encoded["price"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
def model_acc(model):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = model.score(x_test, y_test)
    print(str(model) + "-->" + str(accuracy))
```

```python
lr = LinearRegression()
model_acc(lr)

ls = Lasso()
model_acc(ls)

dt = DecisionTreeRegressor()
model_acc(dt)

rf = RandomForestRegressor()
model_acc(rf)
```

```python
df = pd.DataFrame(columns=["Predicted","Actual"])
df
```

```python
y_pred = rf.predict(x_test)
df["Predicted"] = y_pred
```

```python
params = {"n_estimators":[10,50,100],
         "criterion":["squared_error","absolute_error","poisson"]}
model = RandomForestRegressor()
cval = KFold(n_splits=10)
```

```python
gsearch = GridSearchCV(model, params, cv=cval)
result = gsearch.fit(x_train, y_train)
best_model = result.best_estimator_
best_model
```

```python
best_model.score(x_test, y_test)
```
